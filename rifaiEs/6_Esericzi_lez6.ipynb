### **1. Distribuzioni di probabilità**
Genera 1000 valori casuali da una distribuzione gamma con parametro di forma pari a 1.
Suggerimento: il parametro di forma viene passato come primo argomento quando si "congela" la distribuzione.

Traccia l’istogramma del campione e sovrapponi la PDF della distribuzione.

Stima il parametro di forma dal campione usando il metodo fit.

Extra:
Le distribuzioni hanno molti metodi utili. Esplorali usando il completamento automatico con il tasto TAB.

Traccia la funzione di distribuzione cumulativa (CDF).

Calcola la varianza.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma   

rng = np.random.default_rng(seed=42)        
k, theta = 1.0, 1.0                        
sample = rng.gamma(shape=k, scale=theta, size=1000)

fig, ax = plt.subplots()
ax.hist(sample, bins='auto', density=True, alpha=0.35, label='Campione')
x = np.linspace(0, sample.max(), 400)
ax.plot(x, gamma.pdf(x, a=k, scale=theta), lw=2, label='PDF teorica')
ax.set_xlabel('x'); ax.set_ylabel('densità')
ax.legend(); ax.set_title('Gamma(k=1, θ=1)')
plt.show()
print(f"Media campione: {sample.mean():.2f}")
print(f"Varianza campione: {sample.var(ddof=1):.2f}")
```

### **2. Fitta i dati**
Prova a fittare i dati sottostante con le migliori curve, calcola il MAE e l'RMSE

```python
import numpy as np
temp_max = np.array([17, 19, 21, 28, 33, 38, 37, 37, 31, 23, 19, 18])
temp_min = np.array([-62, -59, -56, -46, -32, -18, -9, -13, -25, -46, -52, -58])
import matplotlib.pyplot as plt
months = np.arange(12)
plt.plot(months, temp_max, "ro")
plt.plot(months, temp_min, "bo")
plt.xlabel("Month")
plt.ylabel("Min and max temperature")
plt.show()
```

```python
from scipy.optimize import curve_fit

def sinusoidal(m, A, phi, B):
    return A * np.sin(2*np.pi*(m + phi)/12) + B

# fit separato per min e max
popt_max, _ = curve_fit(sinusoidal, months, temp_max,
                        p0=[(temp_max.max()-temp_max.min())/2, 0, temp_max.mean()])
popt_min, _ = curve_fit(sinusoidal, months, temp_min,
                        p0=[(temp_min.max()-temp_min.min())/2, 0, temp_min.mean()])
```

```python
from sklearn.metrics import (
    mean_absolute_error,
    root_mean_squared_error   # NEW
)

# 1) calcola le predizioni
pred_max = sinusoidal(months, *popt_max)
pred_min = sinusoidal(months, *popt_min)

# 2) metriche
mae_max  = mean_absolute_error(temp_max, pred_max)
rmse_max = root_mean_squared_error(temp_max, pred_max)   # NEW

mae_min  = mean_absolute_error(temp_min, pred_min)
rmse_min = root_mean_squared_error(temp_min, pred_min)   # NEW

print(f"Max →  MAE {mae_max:.2f}, RMSE {rmse_max:.2f}")
print(f"Min →  MAE {mae_min:.2f}, RMSE {rmse_min:.2f}")
```

```python
#grafico
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(months, temp_max, 'ro', label='max (obs)')
ax.plot(months, pred_max, 'r-', label='max (fit)')
ax.plot(months, temp_min, 'bo', label='min (obs)')
ax.plot(months, pred_min, 'b-', label='min (fit)')
ax.set_xticks(months); ax.set_xlabel('Mese')
ax.set_ylabel('Temperatura [°C]'); ax.legend()
plt.show()
```

### **3. Modello di regressione lineare dei seguenti dati**

 Calcola un modello di regressione lineare delle colonne mpg e disp del famoso dataset mtcars.

Dove: 

mpg = Miles Per Gallon → miglia per gallone, cioè una misura del consumo di carburante. Più alto è il valore, più efficiente è l’auto.

disp = Displacement → cilindrata del motore, in pollici cubici (cubic inches). Rappresenta il volume totale dei cilindri del motore. Più è alto, maggiore è la potenza potenziale del motore (ma anche il consumo).

```python
import pandas as pd

# Load dataset from URL
df = pd.read_csv('https://gist.githubusercontent.com/seankross/a412dfbd88b3db70b74b/raw/5f23f993cd87c283ce766e7ac6b329ee7cc2e1d1/mtcars.csv')

# Display the dataframe
df
```

```python
X = df[['disp']].values
#print(X)
Y = df['mpg'].values

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

model = LinearRegression()
model.fit(X, Y)
coef, intercept = model.coef_[0], model.intercept_
print(f"mpg = {coef:.3f} * disp + {intercept:.3f}")
""" X = df[['disp']].values            
y = df['mpg'].values           

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

model = LinearRegression()          
model.fit(X, y)                  

coef, intercept = model.coef_[0], model.intercept_
print(f"mpg = {coef:.3f} * disp + {intercept:.3f}")
 """
```

```python
y_pred = model.predict(X)            # ⑥
r2 = r2_score(Y, y_pred)             # ⑦
print(f"R² = {r2:.3f}")
```
